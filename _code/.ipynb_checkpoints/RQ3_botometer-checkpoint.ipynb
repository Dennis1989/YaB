```python
import os
import pandas as pd

new_path = os.path.split(os.getcwd())[0]

import json
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

### prepare data

```python
bot_direct = pd.read_pickle(os.path.join(new_path,'_data','bot_direct.pkl'))
```

#### find accounts with min. 2 different accusers

```python
pd_bots = bot_direct[bot_direct['n_accusers']>=2]
pd_bots = pd_bots[(~pd_bots['t1_id'].isnull())&(~pd_bots['t1_author_id'].isnull())]
pd_bots = pd_bots[pd_bots['t1_author_verified']==False]
pd_bots = pd_bots.drop_duplicates('t1_author_id', keep='last')
pd_bots = pd_bots[['t1_author_id','t1_year','n_accusations','n_accusers','n_accusing','n_accusees']]
pd_bots = pd_bots.reset_index(drop=True)
```

```python
len(pd_bots)
```

```python
pd_bots.to_pickle(os.path.join(new_path,'_data','_botometer','botometer_dataset.pkl'))
```

#### augment botometer_dataset with collected botometer scores

```python
pd_bots = pd.read_pickle(os.path.join(new_path,'_data','_botometer','botometer_dataset.pkl'))
```

```python
with open(os.path.join(new_path,'_data','_botometer','botometer_scores.json'),'r') as f:
    botometer_results = json.load(f)
    
botometer = []
for k,v in botometer_results.items():
    try:
        botometer.append([k,v['cap']['english'],v['cap']['universal'],v['raw_scores']['english']['astroturf'],v['raw_scores']['english']['fake_follower'],v['raw_scores']['english']['financial'],v['raw_scores']['english']['other'],v['raw_scores']['english']['overall'],v['raw_scores']['english']['self_declared'],v['raw_scores']['english']['spammer']])
    except:
        botometer.append([k,None,None,None,None,None,None,None,None,None])
        
pd_botometer = pd.DataFrame(botometer, columns=['t1_author_id','t1_author_botometer_cap_english','t1_author_botometer_cap_universal','t1_author_botometer_raw_astroturf','t1_author_botometer_raw_fake_follower','t1_author_botometer_raw_financial','t1_author_botometer_raw_other','t1_author_botometer_raw_overall','t1_author_botometer_raw_self_declared','t1_author_botometer_raw_spammer'])
pd_botometer = pd_botometer[['t1_author_id','t1_author_botometer_raw_overall']]
pd_bots = pd_bots.merge(pd_botometer, on='t1_author_id', how='left')
```

```python
len(pd_bots)
```

```python
pd_bots.to_pickle(os.path.join(new_path,'_data','_botometer','botometer_dataset.pkl'))
```

### generate results

```python
pd_bots = pd.read_pickle(os.path.join(new_path,'_data','_botometer','botometer_dataset.pkl'))
```

#### bot score distribution over time

```python
fig, axs = plt.subplots(2, figsize=(7,4))

## bimodal distributon
axs[0].hist([x for x in pd_bots['t1_author_botometer_raw_overall'] if x <0.5], bins=[i/100 for i in range(101)],edgecolor='black')
axs[0].hist([x for x in pd_bots['t1_author_botometer_raw_overall'] if x >0.5], bins=[i/100 for i in range(101)],edgecolor='black')
axs[0].set_xlabel('botometer score')
axs[0].set_ylabel('frequency')

## scores over time
dic = {}
for year in range(2008,2023):
    pd_bots_filtered = pd_bots[pd_bots['t1_year'] == year]
    dic[year] = pd_bots_filtered['t1_author_botometer_raw_overall'][~np.isnan(pd_bots_filtered['t1_author_botometer_raw_overall'])]

axs[1].boxplot(dic.values() )
axs[1].set_xticklabels(dic.keys())
axs[1].set_ylabel('botometer score')
    
plt.tight_layout()
plt.savefig(os.path.join(new_path,'_results','fig6_botometer_scores.pdf'))
plt.show()
```

#### bot scores x accusations

```python
pd_bots_filtered = pd_bots[~pd_bots['t1_author_botometer_raw_overall'].isnull()]
```

```python
len(pd_bots_filtered)
```

```python
pd_bots_filtered_pre = pd_bots_filtered[(pd_bots_filtered['t1_year']<2017)&(pd_bots_filtered['n_accusers']<40)]
pd_bots_filtered_post = pd_bots_filtered[(pd_bots_filtered['t1_year']>=2017)&(pd_bots_filtered['n_accusers']<40)]
```

```python
fig, axs = plt.subplots(1,2,figsize=(7,2.5))
fig.add_subplot(111, frameon=False)

print(f'correlation before 2017: {len(pd_bots_filtered_pre)} accounts')
print(np.corrcoef(pd_bots_filtered_pre['n_accusers'], pd_bots_filtered_pre['t1_author_botometer_raw_overall']))
print(stats.spearmanr(pd_bots_filtered_pre['n_accusers'], pd_bots_filtered_pre['t1_author_botometer_raw_overall']))
axs[0].scatter(pd_bots_filtered_pre['n_accusers'],pd_bots_filtered_pre['t1_author_botometer_raw_overall'],s=10,marker='+')
axs[0].set_ylabel('Botometer score')
axs[0].set_xlim([0, 40])  

print(f'correlation after 2017: {len(pd_bots_filtered_post)} accounts')
print(np.corrcoef(pd_bots_filtered_post['n_accusers'], pd_bots_filtered_post['t1_author_botometer_raw_overall']))
print(stats.spearmanr(pd_bots_filtered_post['n_accusers'], pd_bots_filtered_post['t1_author_botometer_raw_overall']))
axs[1].scatter(pd_bots_filtered_post['n_accusers'],pd_bots_filtered_post['t1_author_botometer_raw_overall'],s=10,marker='+')
axs[1].set_xlim([0, 40])  

plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
plt.xlabel('Number of accusers')
plt.savefig(os.path.join(new_path,'_results','fig7_accusers_correlation.pdf'), bbox_inches='tight')
plt.show()
```
